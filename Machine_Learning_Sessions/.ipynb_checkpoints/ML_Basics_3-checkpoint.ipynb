- **LogisticRegression** :
Here the model uses the Logistic function (AKA Sigmoid function) instead of a linear one, given by this below equation. It is generally used for *binary classification*.

Here the theta.T * X is nothing but the matrix way of writing. Theta1 * x1 + Theta2 * x2 + ....Thetan * xn + c (Just the dot product of weights and inputs)

The graph looks like follows.
<img src="Images/Logistic.png" alt="Activation functions uses" width="400"/>
Ref: https://towardsdatascience.com/understanding-logistic-regression-step-by-step-704a78be7e0a

Thus if the predicted output is >0.5 then label is 1 and <0.5 label is 0

- **DecisionTreeClassifier**
    <img src="Images/DecisionTree.png" alt="Activation functions uses" width="400"/>
    Ref: https://smritimishra.in/2021/05/13/decision-tree-algorithm-supervised-learning/
    Here we ask questions like: Is the person's salary more than 10,000 rupees? and based on the answer , go down the tree. Each question adds a new node to the tree.
    
    
    As seen above, getting perfect trees are hard. The most common label in the leaf node is taken as the predicted label.
    
    
    This is generally used for classification, but it can be used for regression as well.

- **SVM (Support Vector Machine)**
    
    <img src="Images/SVM.jpg" alt="Activation functions uses" width="700"/>
    
    - Hyper- Plane: The decision boundry that separates two different classes in SVM.
    - Support vectors: The data points that are nearest to the hyper-plane and affect the position and orientation of the hyper-plane.
    
    <img src="Images/SVM_Optimal.png" alt="Activation functions uses" width="700"/>
    The optimal Hyperplane is the one that has maximum distance from the support vectors
    

- Projection into Higher dimensions:
    What if the data looked like this?

    Ref: https://www.analyticsvidhya.com/blog/2021/03/beginners-guide-to-support-vector-machine-svm/
