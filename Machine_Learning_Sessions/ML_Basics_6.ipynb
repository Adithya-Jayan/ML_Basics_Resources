## Convolution Layers:

### How do images work?

- They are made of pixels. The image essentially is a matrix of these pixel values.
- Each pixel is generally represented by a number between 0 and 255 in a black and white image, 0 for full black and 255 for white.
- For colour images, each image has 3 values, corresponding to Red,Green and Blue values.
- The dimensions of the matrix is called the resolution of the image.


(If it is a say, a PNG image that has transparency info as well, then it will have an additional alpha channel besides the RGB )

Before we use the image, we generally normalize the pixel values. Thus we devide the value by 255 and center it at 0.


Say I have a image with a resolution of (1000,1000) as the input.
